```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp2d
from mpl_toolkits.mplot3d import Axes3D
```

```python
#  Our data points in x and y
x = np.linspace(0, 1, 6)
y = np.linspace(0, 1, 6)
X, Y = np.meshgrid(x, y)

#  Our z data points
z = 0.1 * (X**2 + Y**2)

#  Create the interpolator
f = interp2d(X, Y, z)
u = np.linspace(0, 1, 20)
v = np.linspace(0, 1, 20)
U, V = np.meshgrid(u, v)

# Plot the points in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X.flatten(), Y.flatten(), z, 'ro')

ax.plot_surface(U, V, f(u, v), cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
